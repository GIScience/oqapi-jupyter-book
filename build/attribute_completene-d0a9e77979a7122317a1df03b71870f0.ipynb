# Attribute Completeness

The attribute completenss indicator derives the ratio of OSM features that hold one or multiple attributes. Depending on the feature, you can choose from a different set of attributes. If you choose more than one attribute in one querie, the result will be combined and the result figure will show the percentage of all features, that hold all of your chosen attributes.

![attr_comp_menu](images/attribute_comp_01.png)

In this drop down manu you can select one ore multiple attributes.

## Methods
-> Intrinsic approach

Calculates the percentage of features that contain a certain attribute. If the feature you are investigating is from geometry line, the quota is calculate bades on the length in km. For geometry node, the number of all features is compared with the number of features presenting the selected attribute/s.

## Result figure
You find the results of the query in the result log below.
The green, yellow or red banner on the top left indicates the completeness level.

There are three quality categories: 
- low completeness < 25%
- medium completeness 25% - 75%
- high completeness > 75%

The thresholds for the quality categories where selected by viewing the quota distribution of many areas and attributes and must not be seen as an objective metric.
