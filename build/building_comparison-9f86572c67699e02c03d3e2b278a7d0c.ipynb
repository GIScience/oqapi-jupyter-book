# Building Comparison

With the building comparison indicator, you can check the ratio of OSM building area compared to two reference datasets.
It is available only for the bilding area feature.


### Methods and Reference Datasets
-> Extrinsic approach

The result is the ratio of the total area of buildings in OSM devided by the total area of buildings  in the reference dataset. 
According to the availability of the reference dataset in your area of interest, the result log will provide you with one, or both comparison figures.
The building comparison is an extrinsic method, meaning it compares OSM data to non-OSM reference datasets.
This provides and indication for the completeness of OSM data. However, the reference datasets must not be seen as overall accurate and up to date.
The result is the ratio of the total area of buildings in OSM devided by the total area of buildings in the reference dataset.

Reference datasets:
- [EUBUCCO](https://docs.eubucco.com/): EUBUCCO is an europe wide building footprint dataset, derived from administrative datasets. Over 300 million building footprints across europe are covered. This dataset is the most complete dataset for building footprints. However, it is older and could be outdated and it does only cover areas inside of europe.
- [Microsoft Buildings](https://planetarycomputer.microsoft.com/dataset/ms-buildings): The Microsoft Buildings dataset is a worldwide building footprints dataset. A machine learning approach was used to derived building footprints from satellite imagery. Besides the not preventable mistakes that happen with machine learning feature detection, the big advantage of this dataset is the world wide coverage.

In the area of interest panel at the dashboard, you can see white, light grey and dark grey regions, when activating the building comparison indicator. This indicates the availability of the reference datasets. 

- In white areas, both reference datasets are availabe.
- Light grey areas indicate, that only one of the two possiple reference datasets is available.
- Dark grey areas are areas without reference datasets.

A result will be calculated if at least 50% of your selected area is covered by a reference dataset.
