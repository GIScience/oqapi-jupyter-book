# Mapping Saturation

The Mapping Saturation Indicator provides (alongside the completeness indicators) a measure of completeness by analyzing how the
growth of mapped features in OSM approaches a natural maximum over time. 


## Methods and Data
-> Intrinsic approach

The central premise of the mapping saturation is that each feature has a finite extent in reality. As mapping activity increases, the number or length of newly added objects per time step decreases, and the cumulative growth curve gradually saturates toward the true value.
To estimate this saturation level, we use the historical evolution of OSM contributions and fit non-linear regression models, to the cumulative growth curve. The workflow for the mapping saturation indicator consists of four main steps. 

1) The temporal evolution of OSM data is extracted for the selected feature class, aggregated at monthly intervals beginning in 2008.
2) The cumulative time series of mapped features is fitted with different logistic growth models using a non- linear regression approach.
3) The best-fitting model is selected based on statistical error criteria.
4) The asymptote of the fitted model is used to approximate the maximum number of mappable objects, and completeness is calculated as the integral of the actual OSM data to the fitted saturation curve over the last 3 years normalised to
a range of 0 to 1.

This results in three quality steps: 

- Not saturated (<0.3)
- Ongoing mapping (0.3 - 0.97)
- High saturation (> 0.97)

The mapping saturation result curve can be tricky to inperpret. How meaningful the result value is, also comes down to the shape of the saturation curve.

## Limitations
Informativeness varies across topics: performs better for small, discrete objects (e.g., houses) but worse for large polygons (e.g., land use), where a single abrupt mapping step can lead to low completeness.
