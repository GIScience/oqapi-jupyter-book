# Road Comparison

The road comparison indicator predicts the OSM road data completeness based on a reference dataset. The Result is a ratio of the length of reference roads which are covered by OSM roads to the total length of reference dataset.

## Methods and Data
-> Extrinsic method

The indicator identifies corresponding road geometries in OSM and the reference dataset to calculate the ratio of matched road length in km.
The Microsoft Roads dataset is used as a reference. However, the reference dataset must not be seen as overall accurate and up to date.
The road comparison indicator is only available for the cars (all highway) feature.

Reference dataset: 
- [Microsoft Roads](https://github.com/microsoft/RoadDetections): The Microsoft Roads dataset is a worldwide dataset of roads. A machine learning approach was used to derived roads from satellite imagery.
