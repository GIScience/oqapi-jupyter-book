# Attribute Completeness

Der Attribute Completeness-Indikator leitet das Verhältnis von OSM-Features ab, die ein oder mehrere Attribute besitzen. Abhängig vom Feature kannst du aus einer unterschiedlichen Auswahl an Attributen wählen. Wenn du mehr als ein Attribut in einer query auswählst, werden die Ergebnisse kombiniert und die Ergebnisgrafik zeigt den Prozentsatz aller Features, die alle ausgewählten Attribute besitzen.

![attr_comp_menu](../../images/attribute_comp_01.png)

In diesem Drop-down-Menü kannst du ein oder mehrere Attribute auswählen.

## Methodik
-> Intrinsischer Ansatz

Berechnet den Prozentsatz der Features, die ein bestimmtes Attribut enthalten. Wenn das untersuchte Feature eine Liniengeometrie ist, wird die Quote basierend auf der Länge in km berechnet. Für Geometrien vom Typ node wird die Gesamtanzahl aller Features mit der Anzahl der Features verglichen, die die ausgewählten Attribute aufweisen.

## Result figure
Die Ergebnisse der query findest du im Ergebnisprotokoll unten.
Das grüne, gelbe oder rote Banner oben links zeigt den Completeness-Level an.

Es gibt drei Qualitätskategorien:
- niedrige Completeness < 25%
- mittlere Completeness 25% - 75%
- hohe Completeness > 75%

Die Schwellenwerte für die Qualitätskategorien wurden durch die Betrachtung der Quotenverteilung vieler Gebiete und Attribute festgelegt und sollten nicht als objektive Metrik verstanden werden.
