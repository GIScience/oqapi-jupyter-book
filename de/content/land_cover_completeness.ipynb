# Land Cover Completeness

Der Land Cover Completeness-Indikator berechnet den Prozentsatz der Gesamtfläche, die durch OpenStreetMap-Land-Cover-Daten abgedeckt ist.

### Methodik und Daten
-> Intrinsischer Ansatz

Der Land Cover Completeness-Indikator quantifiziert den Anteil eines Interessensgebiets (AOI), der durch OSM-Land-Cover-Daten abgedeckt ist. Er wird intrinsisch als Verhältnis der aufsummierten Fläche von OSM-Land-Cover-Polygonen zur Gesamtfläche des AOI berechnet. Der Ergebniswert wird als Verhältnis der durch LULC abgedeckten Fläche berechnet, normalisiert auf einen Bereich von 0 bis 1:

- Niedrige Completeness (< 0.5)
- Mittlere Completeness (0.5 - 0.85)
- Hohe Completeness (> 0.85)

Dieser Indikator ist nur für den land cover und land use Filter verfügbar. Dieser umfasst alle OSM-Tags, die Landbedeckung und Landnutzung definieren, wie bebaute Flächen, Wald oder Binnengewässer. Eine detaillierte Liste der Filter für alle Features findest du in den [Topic Definitions](Topic_Definitions.md)

### Limitationen

Sich überlappende OSM-Land-Cover-Polygone werden mehrfach gezählt und können das Land-Cover-Completeness-Verhältnis fälschlicherweise erhöhen.
