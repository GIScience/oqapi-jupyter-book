# Currentness

The Currentness Indicator measures how up-to-date OSM elements are by analyzing the distribution of their most recent contributions.

## Methods
-> Intrinsic approach

The currentness indicator considers all edits to geometry and tags (excluding deletions) since 2008, aggregated in monthly intervals. The result value is calculated as the percentage of elements that were last edited in the last 36 months normalised to a range of 0 to 1. Each contribution is assigned to one of three quality levels:

- Outdated (< 0.5): Features that have not been updated for a longer period are labelled as out-of-date. These thresholds are expressed in months since the current date. For each feature class, the relative share of contributions falling into these bins is computed.
- In-between (0.5 - 0.75): Contributions that were last edited within an intermediate range are classified as in-between.
- Up-to-date (> 0.75): Features are considered up-to-date if
their last edit falls within a predefined short time window.
