# Land Cover Completeness

The land cover completeness indicator calculates the percentage of total area that is covered by OpenStreetMap land cover data.

### Methods and Data
-> Intrinsic approach

The land cover completeness indicator quantifies the share
of an area of interest (AOI) covered by OSM land cover
data. It is calculated intrinsically as the ratio of the
summed area of OSM land cover polygons to the total area
of the AOI. The result value is calculated as the ratio of
area covered by LULC normalised to a range of 0 to 1:

- Low completeness (< 0.5)
- Medium completeness (0.5 - 0.85)
- High completeness (> 0.85)

This indicator is only available for the land cover and land use filter. This selects all OSM tags that define land cover and land use like build-up areas, forest or inland waters. You can find a detailed list of filters for all features in the [Topic Definitions](Topic_Definitions.md)

### Limitations

Overlapping OSM land cover polygons will be counted multiple times and falsely improve the land cover completeness ratio.
