# User Activity

This indicator calculates how many unique users contributed to a specific topic, grouped by months.


### Methods and Data
-> Intrinsic method

The monthly amount of unique users who edited a specific topic in the area of interest are derived for the entire time range of OSM.
Additionally, the median for the last 3 years are calculated as well as a regression line to see the current tendency of user activity.

The user activity assesses contributor engagement within the study area and is not directly giving information about data quality.
However, combining it with other indicators and viewing absolute numbers and regression trends allow some conclusions about the sustainability of the OSM community.
